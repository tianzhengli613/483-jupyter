# 1. Download files
***
The data that will be used is `bank-additional.csv`

# 2. read_csv()
***
The data is read, with `;` as the separator

```python
import pandas as pd

# Read data
data = pd.read_csv('bank-additional.csv', sep = ';')
display(data)
```

# 3. Preprocessing data
***
The data is turned into a dummy, and the `duration` and `y_yes` columns are removed

```python
# Convert to dummy
data_d = pd.get_dummies(data)

# Remove unnecessary columns
data_d = data_d.drop('duration', 1)

target = []
for i in range(len(data_d)):
    target.append(data_d['y_yes'][i])
    
data_d = data_d.drop('y_yes', 1)

display(data_d)
```

# 4. Classifiers
***
SVM had the highest accuracy for both training and test set

```python
import numpy as np
from sklearn.model_selection import train_test_split

# Split into training and test data
X_train, X_test = train_test_split(data_d, test_size = 0.2)
y_train, y_test = train_test_split(target, test_size = 0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

# Training
# ===================================
# GNB
train_gnb = GaussianNB()
train_gnb.fit(X_train, y_train)
print("Training set")
print("=================")
print("GNB: " + str(train_gnb.score(X_train, y_train)))

# KNN
train_knn = KNeighborsClassifier()
train_knn.fit(X_train, y_train)
print("KNN: " + str(train_knn.score(X_train, y_train)))

# SVM
train_svm = make_pipeline(StandardScaler(), SVC(gamma = 'auto'))
train_svm.fit(X_train, y_train)
print("SVM: " + str(train_svm.score(X_train, y_train)))

# Testing
# ===================================
# GNB
test_gnb = GaussianNB()
test_gnb.fit(X_test, y_test)
print("\nTesting set")
print("=================")
print("GNB: " + str(test_gnb.score(X_test, y_test)))

# KNN
test_knn = KNeighborsClassifier()
test_knn.fit(X_test, y_test)
print("KNN: " + str(test_knn.score(X_test, y_test)))

# SVM
test_svm = make_pipeline(StandardScaler(), SVC(gamma = 'auto'))
test_svm.fit(X_test, y_test)
print("SVM: " + str(test_svm.score(X_test, y_test)))
```

# 5. Target values
***
In the training set, the target (`y_yes`) contains almost 90% 0s, meaning most people do not subscribe to the term deposit. If no customers subscribed then the accuracy would be much lower because we would not know how to classify those that would subscribe.

```python
y_yes0 = 0
y_yes1 = 0
for i in range(len(y_train)):
    if y_train[i] == 0:
        y_yes0 += 1
    else:
        y_yes1 += 1
        
print("In a training set of length " + str(len(y_train)) + 
      " there are [" + str(y_yes0) + "] 0s and [" + str(y_yes1) + "] 1s")    
```

# 6. AUC
***
The AUC of the training data with `LogisticRegression` is `0.538`, which was not unable to calculated with just a single class (dummy with just 0s).

```python
from sklearn.metrics import confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score

dumb_train = np.zeros_like(y_train)
dumb_matrix = confusion_matrix(y_train, dumb_train)
print(dumb_matrix)

lr = LogisticRegression(solver="liblinear", random_state=0).fit(X_train, y_train)
print("AUC: " + str(roc_auc_score(y_train, lr.decision_function(X_train))))
```

# 7. AUC for all classifiers
***
The AUC for all 3 of the classifiers (SVM does not work with only 1 class of just 0s) shwos that the they are all roughly the same. The best classifier (SVM) does have the best accuracy but that may not be necessarily the case.

```python
# GNB
dumb_gnb = GaussianNB()
dumb_gnb.fit(X_train, dumb_train)
print("GNB AUC: " + str(roc_auc_score(y_train, dumb_gnb.predict(X_train))))
print(confusion_matrix(y_train, dumb_gnb.predict(X_train)))
print()

# KNN
dumb_knn = KNeighborsClassifier()
dumb_knn.fit(X_train, dumb_train)
print("KNN AUC: " + str(roc_auc_score(y_train, dumb_knn.predict(X_train))))
print(confusion_matrix(y_train, dumb_knn.predict(X_train)))
print()

# SVM
dumb_svm = make_pipeline(StandardScaler(), SVC(gamma = 'auto'))
dumb_svm.fit(X_train, y_train)
print("SVM AUC: " + str(roc_auc_score(y_train, dumb_svm.predict(X_train))))
print(confusion_matrix(y_train, dumb_svm.predict(X_train)))
```

# 8. Random Oversampling
***
